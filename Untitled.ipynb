```python
import pandas as pd
import openpyxl
```

```python
# Read file, set index, transform and rename odd column
df = pd.read_excel('202221172517361789673SMIT1.xlsx', skiprows=2)
df.set_index('Unnamed: 0', inplace=True)
df = df.T
df.columns = ['cumulative_infected']
```

```python
# Remove the days with missing values (20 days total) and convert to int
df = df[df['cumulative_infected'] != '..']
df['c'] = df['cumulative_infected'].astype(int)
df.drop(labels=['cumulative_infected'], axis=1, inplace=True)
```

```python
# Create a column where we get differences
df['diff'] = df['c'].diff()

# Make new columns for one week and two week rolling sums
df['one_week'] = df['diff'].rolling(min_periods=1, window=7).sum()
df['two_week'] = df['diff'].rolling(min_periods=1, window=14).sum()
```

```python
# Dump to Excel
df.to_excel('output.xlsx')
```
