```python
import pandas as pd
import numpy as np
import openpyxl
```

```python
# Read file, set index, transform and rename odd column
df = pd.read_excel('202222162642361881182SMIT1.xlsx', skiprows=2, skipfooter=2)
df.set_index('Unnamed: 0', inplace=True)
df = df.T
df.columns = ['cum_infected', 'hospital','hospital_intensive', 'vaccine_init', 'vaccine_finished']
```

```python
# Converts the ".." to np.NaN
for col in list(df.columns):
    df[col] = df[col].apply(lambda x: np.NaN if x == '..' else x)
```

```python
# Unwraps the cumulative infected
df['infected'] = df['cum_infected'].diff()

# Gets vaccine percentage
df['vaccine_percent'] = df['vaccine_finished'].apply(lambda x: round(x / 5800000 * 100, 2))
```

```python
# Make new columns for one week and two week rolling sums
df['one_week'] = df['infected'].rolling(min_periods=1, window=7).sum()
df['two_week'] = df['infected'].rolling(min_periods=1, window=14).sum()
```
