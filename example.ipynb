```python
import pandas as pd
import numpy as np
```

```python
# Reading an excel spreadsheet, taking the first sheet as a default
# Skipping the first two rows setting headers to row 2 (3 in Excel being 1 based)
df_users = pd.read_excel('source_name.xlsx', header=2)

# Setting the index and doing an inplace replacement
df_users.set_index('Nummer', inplace=True)

# Printing the result
df_users
```

```python
# Reading an excel spreadsheet, specifying the sheet name and header row
df_call_records = pd.read_excel('sources.xlsx', sheet_name='source_1', header=0)

# printing the results
df_call_records
```

```python
# Making an outer join, specifying the first dataframe merging the next into.
# How specifies the method (inner, outer, left etc), and you can specify if it 
# should be the index or a specific column
df_result = df_users.merge(df_call_records, how='outer', left_index=True, right_on='ID')

# Drops an unneccessary column
df_result.drop('ID', inplace=True, axis=1)

# printing the result
df_result
```

```python
# Reads an excel file and specifies that there are no headers
df_random = pd.read_excel('sources.xlsx', sheet_name='source_2', header=None)

# Adding my own column values
df_random.columns = ['id','value']

# Printing the result
df_random
```

```python
# Gets duplicated values, can also be set to keep only the first or the last value...
df_random[df_random.duplicated(subset=['id'], keep=False)]
```

```python
df_random[df_random.duplicated(subset=['id'], keep='first')]
```

```python
df_random[df_random.duplicated(subset=['id'], keep='last')]
```
